```python
import matplotlib.pyplot as plt  # matplotlib
import qibo  # qibo
from qibo import gates, models  # qibogatesmodels
from qibo.models import QFT  # qibo.modelsQFT

# matplotlib
from qibo.ui import plot_circuit  # qibo.uiplot_circuit

# 5QFT
circuit = QFT(5)
circuit.add(gates.M(qubit) for qubit in range(2))  # 

# 0.6
plot_circuit(circuit);

# "garnacha"
# "garnacha", "fardelejo", "quantumspain", "color-blind", "cachirulo" 
plot_circuit(circuit, scale=0.8, cluster_gates=True, style="garnacha");

# Qibo
custom_style = {
    "facecolor": "#6497bf",  # 
    "edgecolor": "#01016f",  # 
    "linecolor": "#01016f",  # 
    "textcolor": "#01016f",  # 
    "fillcolor": "#ffb9b9",  # 
    "gatecolor": "#d8031c",  # 
    "controlcolor": "#360000"  # 
}

plot_circuit(circuit, scale=0.8, cluster_gates=True, style=custom_style);
```
